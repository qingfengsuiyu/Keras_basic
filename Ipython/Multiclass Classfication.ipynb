```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Activation
from keras.utils.np_utils import to_categorical
from keras.utils import plot_model as plot
```

```python
model = Sequential()
model.add(Dense(32,input_dim=500))
model.add(Activation(activation='relu'))
model.add(Dense(10))
model.add(Activation(activation='softmax'))
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['categorical_accuracy'])
```

```python
data = np.random.random((1000,500))
lables = to_categorical(np.random.randint(10,size=(1000,1)))
```

```python
score = model.evaluate(data,lables,verbose=0)
print("Before Training:",list(zip(model.metrics_names,score)))
```

```python
model.fit(data,lables,nb_epoch=10,batch_size=32,verbose=0)
```

```python
score = model.evaluate(data,lables,verbose=0)
print("After Training:",list(zip(model.metrics_names,score)))
```

```python
plot(model,to_file='s3.png',show_shapes=True)
```
